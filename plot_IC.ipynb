```python
# necessary imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# populate this cell once we know the format of the input data
# sep lets us read in a tab-separated textfile as a csv
infile = pd.read_csv("../../Flat_REBCO_Data_Fall25/12-05-2025/6mm-eb-30mm-run2.txt", sep = "\t")
print(infile.columns)
```

```python
# fill these with the input data
V_tot = np.array(infile['CH3'])
V_tape = np.array(infile['CH5']) # should I be averaging channel 5 and channel 6??
I_input = np.array(infile['CURRENT'])

# measured/calculated value of cable resistance
R_cable = 5.8e-5
```

```python
I_cable = V_tot / R_cable
I_tape = I_input - I_cable
```

```python
plt.scatter(I_tape, V_tape, marker = '.', color = 'red')
plt.xlabel("$I_{tape}$ [A]")
plt.ylabel("$V_{tape}$ [V]")
```

```python
plt.scatter(I_input, V_tot, marker = '.', color = 'red')
plt.xlabel("$I_{cable}$ [A]")
plt.ylabel("$V_{cable}$ [V]")
```

```python
plt.scatter(I_input, V_tape, marker = '.', color = 'red', label = "Measured (total)")
plt.scatter(I_tape, V_tape, marker = '.', color = 'blue', label = "Tape Only")
plt.xlabel("$I$ [A]")
plt.ylabel("$V$ [V]")
plt.legend()
```
